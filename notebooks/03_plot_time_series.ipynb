# PySWEB example: plot SSEBop + SWB time series

This notebook runs `python -m pysweb.visualisation.plot_time_series` using explicit SSEBop and SWB NetCDF paths.


## Configure paths and run settings

Set point coordinates for nearest-grid-cell extraction, or omit `--lat` and `--lon` for domain-mean extraction.


```python
from pathlib import Path
import subprocess
import sys

PYSWEB_DIR = Path("/path/to/PySWEB")
RUN_ROOT = Path("/path/to/notebook_runs")
RUN_NAME = "example_run_name"
RUN_DIR = RUN_ROOT / RUN_NAME

SSEBOP_START_DATE = "YYYY-MM-DD"
SSEBOP_END_DATE = "YYYY-MM-DD"
SWB_START_DATE = "YYYY-MM-DD"
SWB_END_DATE = "YYYY-MM-DD"

SSEBOP_PATH = RUN_DIR / "outputs" / "ssebop" / f"et_daily_ssebop_{SSEBOP_START_DATE}_{SSEBOP_END_DATE}.nc"
SWB_PATH = RUN_DIR / "outputs" / "sweb" / f"SWEB_RZSM_{SWB_START_DATE}_{SWB_END_DATE}.nc"

USE_POINT_MODE = True
POINT_LAT = 0.0
POINT_LON = 0.0

OUTPUT_DIR = RUN_DIR / "figures"
OUTPUT_DIR.mkdir(parents=True, exist_ok=True)

TIMESERIES_PNG = OUTPUT_DIR / f"{RUN_NAME}_timeseries_notebook.png"
TIMESERIES_CSV = OUTPUT_DIR / f"{RUN_NAME}_timeseries_notebook.csv"

PLOT_MODULE = "pysweb.visualisation.plot_time_series"
PLOT_MODULE
```

```python
cmd = [
    sys.executable,
    "-m", PLOT_MODULE,
    "--ssebop-path", str(SSEBOP_PATH),
    "--sweb-path", str(SWB_PATH),
    "--start-date", SWB_START_DATE,
    "--end-date", SWB_END_DATE,
    "--output", str(TIMESERIES_PNG),
    "--csv-out", str(TIMESERIES_CSV),
]

if USE_POINT_MODE:
    cmd.extend(["--lat", str(POINT_LAT), "--lon", str(POINT_LON)])

print("Running:")
print(" ".join(cmd))

subprocess.run(cmd, cwd=str(PYSWEB_DIR), check=True)
print(f"Wrote figure: {TIMESERIES_PNG}")
print(f"Wrote CSV:    {TIMESERIES_CSV}")
```

```python
from IPython.display import Image, display
import pandas as pd

display(Image(filename=str(TIMESERIES_PNG)))
pd.read_csv(TIMESERIES_CSV).head()
```
