# PySWEB example: plot point-based SWB heatmap

This notebook runs `python -m pysweb.visualisation.plot_heatmap` using explicit SWB and optional SSEBop NetCDF paths.


## Configure paths and run settings

Update the run metadata, point coordinates, date range, and output location to match a completed `01_run_pysweb.ipynb` run.


```python
from pathlib import Path
import subprocess
import sys

PYSWEB_DIR = Path("/path/to/PySWEB")
RUN_ROOT = Path("/path/to/notebook_runs")
RUN_NAME = "example_run_name"
RUN_DIR = RUN_ROOT / RUN_NAME

SSEBOP_START_DATE = "YYYY-MM-DD"
SSEBOP_END_DATE = "YYYY-MM-DD"
SWB_START_DATE = "YYYY-MM-DD"
SWB_END_DATE = "YYYY-MM-DD"

SWB_PATH = RUN_DIR / "outputs" / "sweb" / f"SWEB_RZSM_{SWB_START_DATE}_{SWB_END_DATE}.nc"
SSEBOP_PATH = RUN_DIR / "outputs" / "ssebop" / f"et_daily_ssebop_{SSEBOP_START_DATE}_{SSEBOP_END_DATE}.nc"

POINT_LAT = 0.0
POINT_LON = 0.0

OUTPUT_DIR = RUN_DIR / "figures"
OUTPUT_DIR.mkdir(parents=True, exist_ok=True)

HEATMAP_PNG = OUTPUT_DIR / f"{RUN_NAME}_heatmap_point_notebook.png"
HEATMAP_CSV = OUTPUT_DIR / f"{RUN_NAME}_heatmap_point_notebook.csv"

PLOT_MODULE = "pysweb.visualisation.plot_heatmap"
PLOT_MODULE
```

```python
cmd = [
    sys.executable,
    "-m", PLOT_MODULE,
    "--sweb-path", str(SWB_PATH),
    "--ssebop-path", str(SSEBOP_PATH),
    "--lat", str(POINT_LAT),
    "--lon", str(POINT_LON),
    "--start-date", SWB_START_DATE,
    "--end-date", SWB_END_DATE,
    "--vmin", "0.1",
    "--vmax", "0.6",
    "--output", str(HEATMAP_PNG),
    "--csv-out", str(HEATMAP_CSV),
]

print("Running:")
print(" ".join(cmd))

subprocess.run(cmd, cwd=str(PYSWEB_DIR), check=True)
print(f"Wrote figure: {HEATMAP_PNG}")
print(f"Wrote CSV:    {HEATMAP_CSV}")
```

```python
from IPython.display import Image, display
import pandas as pd

display(Image(filename=str(HEATMAP_PNG)))
pd.read_csv(HEATMAP_CSV).head()
```
